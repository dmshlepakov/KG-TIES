```
df1 = dfs['1_population']
```

```
df1['Study Link']
```

```
df1['doi'] = df1.apply(lambda x: x['Study Link'].replace('https://doi.org/','') \
                       .replace('http://doi.org/', '') \
                       .replace('https://dx.doi.org/','') \
                       .replace('http://dx.doi.org/',''), axis =1 )
```

```
df1['pmcid'] = df1.apply(lambda x: x['Study Link'] \
                       .replace('https://www.ncbi.nlm.nih.gov/pmc/articles/','') \
                       .replace('http://www.ncbi.nlm.nih.gov/pmc/articles/','') \
                       .replace('/',''), axis =1 )
```

```
df1['arxiv_id'] = df1.apply(lambda x: x['Study Link'] \
                       .replace('https://arxiv.org/pdf/','') \
                       .replace('http://arxiv.org/pdf/','') \
                       .replace('.pdf',''), axis =1 )
```

```
meta_data = pd.read_csv('/content/gdrive/My Drive/code/coronawhy-kaggle/metadata.csv',low_memory=False)
```

```
meta_data.head(30)
```

```
meta_data_doi= meta_data[['cord_uid','doi']].set_index('doi')
meta_data_pmcid = meta_data[['cord_uid','pmcid']].set_index('pmcid')
meta_data_pubmed_id = meta_data[['cord_uid','pubmed_id']].set_index('pubmed_id')
meta_data_arxiv_id= meta_data[['cord_uid','arxiv_id']].dropna().set_index('arxiv_id')
```

```
meta_data_doi_dict = meta_data_doi.to_dict()
meta_data_pmcid_dict = meta_data_pmcid.to_dict()
meta_data_pubmed_id_dict = meta_data_pubmed_id.to_dict()
meta_data_arxiv_id_dict = meta_data_arxiv_id.to_dict()
```

```
'10.1186/1471-2334-1-6' in meta_data_doi['cord_uid']
```

```
df1['cord_uid_doi'] = df1.apply(lambda x: meta_data_doi_dict['cord_uid'][x['doi']] if x['doi'] in meta_data_doi_dict['cord_uid'] else None, axis = 1)
df1['cord_uid_pmcid'] = df1.apply(lambda x: meta_data_pmcid_dict['cord_uid'][x['pmcid']] if x['pmcid'] in meta_data_pmcid_dict['cord_uid'] else None, axis = 1)
df1['cord_uid_arxiv_id'] = df1.apply(lambda x: meta_data_arxiv_id_dict['cord_uid'][x['arxiv_id']] if x['arxiv_id'] in meta_data_arxiv_id_dict['cord_uid'] else None, axis = 1)
```
